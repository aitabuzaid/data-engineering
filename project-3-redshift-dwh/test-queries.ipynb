```python
%%sql 
SELECT songplays.user_id, users.first_name, users.last_name, users.level, COUNT(songplays.user_id)
FROM songplays 
JOIN users ON songplays.user_id = users.user_id
GROUP BY songplays.user_id, users.first_name, users.last_name, users.level
ORDER BY count DESC
```

```python
%%time
%%sql 
SELECT level, COUNT(user_id) 
FROM users 
GROUP BY level
```
